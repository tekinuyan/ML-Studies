```python
def parsty(data):
    tree = ET.parse(data)
    root = tree.getroot()
    name=[]
    for elm in root.findall(".//ForeName"):
        name.append(elm.text)
    lastname=[]
    for elm in root.findall(".//LastName"):
        lastname.append(elm.text)
    affili=[]
    for elm in root.findall(".//Affiliation"):
        affili.append(elm.text)  
    country=[i.split(',')[-1] for i in affili]
    country=[i.split('.') for i in country]
    b=[]
    for i in range(len(country)):
        b.append(country[i][0]) 
    country==b
    df = pd.DataFrame(list(zip(name, lastname,country)),
               columns =['Name', 'LastName','country'])
    return df
```
